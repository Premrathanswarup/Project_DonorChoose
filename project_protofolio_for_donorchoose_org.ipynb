```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
#	Project Essay Length
# Load the dataset
df = pd.read_csv('train_data.csv')

# Combine the essay parts into a single essay
df['full_essay'] = (df['project_essay_1'].fillna('') + ' ' +
                    df['project_essay_2'].fillna('') + ' ' +
                    df['project_essay_3'].fillna('') + ' ' +
                    df['project_essay_4'].fillna(''))

# Calculate the word count for each essay
df['essay_length'] = df['full_essay'].apply(lambda x: len(x.split()))

# Plotting the histogram of essay lengths
plt.figure(figsize=(10, 6))
sns.histplot(df['essay_length'], bins=30, kde=True)
plt.xlabel('Essay Length (Word Count)')
plt.ylabel('Frequency')
plt.title('Distribution of Essay Lengths')
plt.show()

# Plotting the box plot of essay lengths to identify outliers
plt.figure(figsize=(10, 6))
sns.boxplot(df['essay_length'])
plt.xlabel('Essay Length (Word Count)')
plt.title('Box Plot of Essay Lengths')
plt.show()

# Identifying outliers using IQR method
Q1 = df['essay_length'].quantile(0.25)
Q3 = df['essay_length'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['essay_length'] < (Q1 - 1.5 * IQR)) | (df['essay_length'] > (Q3 + 1.5 * IQR))]
print(f"Number of outliers: {outliers.shape[0]}")
print(outliers[['id', 'essay_length']])

# Plotting the box plot of essay lengths by approval status
plt.figure(figsize=(10, 6))
sns.boxplot(x='project_is_approved', y='essay_length', data=df)
plt.xlabel('Approval Status')
plt.ylabel('Essay Length (Word Count)')
plt.xticks([0, 1], ['Rejected', 'Approved'])
plt.title('Essay Length by Approval Status')
plt.show()

# Calculate and print the correlation coefficient
correlation = df['project_is_approved'].corr(df['essay_length'])
print(f'Correlation between Approval Status and Essay Length: {correlation:.2f}')

# Perform a statistical test to see if there's a significant difference in essay length based on approval status
approved = df[df['project_is_approved'] == 1]['essay_length']
rejected = df[df['project_is_approved'] == 0]['essay_length']

t_stat, p_val = ttest_ind(approved, rejected)
print(f'T-test: t_stat = {t_stat}, p_val = {p_val}')

#Correlation Analysis:

#The correlation coefficient between essay_length and project_is_approved is 0.07, which is very close to zero. This indicates that there is a weak positive linear relationship between essay length and approval rates. Although the correlation is weak, it is still positive, suggesting that longer essays might be slightly more likely to be approved.
#T-Test:

#The t-test result shows a t_stat of 23.92 and a p_val of 4.31e-126. The extremely low p-value indicates that the difference in the mean essay lengths of approved and rejected projects is statistically significant. Despite the weak correlation, the significant p-value suggests that essay length does have some impact on approval likelihood.#
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

# Load the data
train_data = pd.read_csv('train_data.csv')
resources = pd.read_csv('resources.csv')

# Calculate total project cost
resources['total_cost'] = resources['price'] * resources['quantity']
project_costs = resources.groupby('id')['total_cost'].sum().reset_index()

# Merge project costs with train_data
train_data = train_data.merge(project_costs, on='id', how='left')

# Plot distribution of project costs
plt.figure(figsize=(10, 6))
sns.histplot(train_data['total_cost'], bins=50, kde=True)
plt.title('Distribution of Total Project Costs')
plt.xlabel('Total Project Cost')
plt.ylabel('Frequency')
plt.show()

# Calculate skewness and kurtosis
original_skewness = skew(train_data['total_cost'].dropna())
original_kurtosis = kurtosis(train_data['total_cost'].dropna())
print(f"Original Skewness: {original_skewness}")
print(f"Original Kurtosis: {original_kurtosis}")

# Identify outliers using the IQR method
Q1 = train_data['total_cost'].quantile(0.25)
Q3 = train_data['total_cost'].quantile(0.75)
IQR = Q3 - Q1

# Define outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = train_data[(train_data['total_cost'] < lower_bound) | (train_data['total_cost'] > upper_bound)]

print(f"Number of outliers: {len(outliers)}")

# Winsorize the data
train_data['total_cost_winsorized'] = np.where(train_data['total_cost'] < lower_bound, lower_bound,
                                               np.where(train_data['total_cost'] > upper_bound, upper_bound, train_data['total_cost']))

# Remove outliers
train_data_no_outliers = train_data[(train_data['total_cost'] >= lower_bound) & (train_data['total_cost'] <= upper_bound)]

# Calculate skewness and kurtosis after winsorization
winsorized_skewness = skew(train_data['total_cost_winsorized'].dropna())
winsorized_kurtosis = kurtosis(train_data['total_cost_winsorized'].dropna())
print(f"Winsorized Skewness: {winsorized_skewness}")
print(f"Winsorized Kurtosis: {winsorized_kurtosis}")

# Calculate skewness and kurtosis after removing outliers
no_outliers_skewness = skew(train_data_no_outliers['total_cost'].dropna())
no_outliers_kurtosis = kurtosis(train_data_no_outliers['total_cost'].dropna())
print(f"No Outliers Skewness: {no_outliers_skewness}")
print(f"No Outliers Kurtosis: {no_outliers_kurtosis}")

# Plot distribution of winsorized project costs
plt.figure(figsize=(10, 6))
sns.histplot(train_data['total_cost_winsorized'], bins=50, kde=True)
plt.title('Distribution of Winsorized Total Project Costs')
plt.xlabel('Total Project Cost (Winsorized)')
plt.ylabel('Frequency')
plt.show()

# Plot distribution of project costs without outliers
plt.figure(figsize=(10, 6))
sns.histplot(train_data_no_outliers['total_cost'], bins=50, kde=True)
plt.title('Distribution of Total Project Costs (Without Outliers)')
plt.xlabel('Total Project Cost')
plt.ylabel('Frequency')
plt.show()
print(train_data['total_cost_winsorized'].describe())
print("=====")
print(train_data_no_outliers['total_cost'].describe())

#Both winsorization and outlier removal are effective in handling the skewed data
#The mean and median are slightly higher in the winsorized data compared to the outlier-removed data,
#The standard deviation is higher in the winsorized data ($349.27) compared to the outlier-removed data ($278.42), reflecting a broader range of values.
#Both methods have similar min and max values due to the winsorization process capping the values at the calculated thresholds.
#The percentiles (25th, 50th, and 75th) are also quite close, showing that both methods effectively handle the central tendencies of the data.

```

```
import pandas as pd
import scipy.stats as stats
##by chi-square test on Project Category on Approval
# Load the data
train_data = pd.read_csv('train_data.csv')

# Extract relevant columns
data = train_data[['project_subject_categories', 'project_is_approved']]

# Handle missing values if any
data.dropna(subset=['project_subject_categories', 'project_is_approved'], inplace=True)

# Create a contingency table
contingency_table = pd.crosstab(data['project_subject_categories'], data['project_is_approved'])
print("Contingency Table:")
print(contingency_table)

# Perform Chi-square test
chi2, p, dof, ex = stats.chi2_contingency(contingency_table)
print(f"Chi-square statistic: {chi2}")
print(f"p-value: {p}")
print(f"Degrees of freedom: {dof}")

# Interpret the results
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: There is a significant association between project category and approval status.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between project category and approval status.")


  #by chi-square test on Project Category on Approval we reject null hypothesis
```

```
import pandas as pd
import scipy.stats as stats
#t-test on Teacher Experience and Approval Rates
# Load the data
train_data = pd.read_csv('train_data.csv')

# Extract relevant columns
data = train_data[['teacher_number_of_previously_posted_projects', 'project_is_approved']]

# Handle missing values if any
data.dropna(subset=['teacher_number_of_previously_posted_projects', 'project_is_approved'], inplace=True)

# Create two groups
first_time_teachers = data[data['teacher_number_of_previously_posted_projects'] == 0]['project_is_approved']
experienced_teachers = data[data['teacher_number_of_previously_posted_projects'] > 0]['project_is_approved']

# Calculate mean approval rates
mean_first_time = first_time_teachers.mean()
mean_experienced = experienced_teachers.mean()
print(f"Mean approval rate for first-time teachers: {mean_first_time}")
print(f"Mean approval rate for experienced teachers: {mean_experienced}")

# Perform t-test
t_stat, p_val = stats.ttest_ind(first_time_teachers, experienced_teachers)
print(f"t-statistic: {t_stat}")
print(f"p-value: {p_val}")

# Interpret the results
alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis: There is a significant difference in approval rates between first-time teachers and experienced teachers.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in approval rates between first-time teachers and experienced teachers.")

#By t-test on Teacher Experience and Approval Rates we can Reject the null hypothesis
```

```
import pandas as pd
import scipy.stats as stats
#	Project Cost and Approval Likelihood:
# Load the data
train_data = pd.read_csv('train_data.csv')
resources = pd.read_csv('resources.csv')

# Calculate total project cost
resources['total_cost'] = resources['price'] * resources['quantity']
project_costs = resources.groupby('id')['total_cost'].sum().reset_index()

# Merge project costs with train_data
train_data = train_data.merge(project_costs, on='id', how='left')

# Drop rows with missing values in the relevant columns
train_data.dropna(subset=['total_cost', 'project_is_approved'], inplace=True)

# Calculate Pearson correlation coefficient
correlation, p_value = stats.pearsonr(train_data['total_cost'], train_data['project_is_approved'])
print(f"Pearson correlation coefficient: {correlation}")
print(f"p-value: {p_value}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant correlation between project cost and approval likelihood.")
else:
    print("Fail to reject the null hypothesis: There is no significant correlation between project cost and approval likelihood.")

# p-value: 1.8518156078432854e-28
#Reject the null hypothesis: There is a significant correlation between project cost and approval likelihood
```

```
import pandas as pd
import scipy.stats as stats

# Load the data
train_data = pd.read_csv('train_data.csv')

# Function to calculate the length of each essay
def calculate_essay_length(row):
    total_length = 0
    for i in range(1, 5):  # There are 4 essays
        essay_col = f'project_essay_{i}'
        if pd.notnull(row[essay_col]):
            total_length += len(row[essay_col])
    return total_length

# Apply the function to calculate essay length for each project
train_data['essay_length'] = train_data.apply(calculate_essay_length, axis=1)

# Create two groups: Approved and Rejected
approved_essays = train_data[train_data['project_is_approved'] == 1]['essay_length']
rejected_essays = train_data[train_data['project_is_approved'] == 0]['essay_length']

# Calculate mean essay lengths
mean_approved = approved_essays.mean()
mean_rejected = rejected_essays.mean()
print(f"Mean essay length for approved projects: {mean_approved}")
print(f"Mean essay length for rejected projects: {mean_rejected}")

# Perform t-test
t_stat, p_val = stats.ttest_ind(approved_essays, rejected_essays, equal_var=False)
print(f"t-statistic: {t_stat}")
print(f"p-value: {p_val}")

# Interpret the results
alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis: There is a significant difference in essay length between approved and rejected projects.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in essay length between approved and rejected projects.")

#Mean essay length for approved projects: 1542.055282290251
#Mean essay length for rejected projects: 1465.9390642002177
#t-statistic: 24.103999290829922
#p-value: 7.125152438830448e-127
#T=test on Impact of Essay Length on Approval we can Reject the null hypothesis: There is a significant difference in essay length between approved and rejected projects
```

```
import pandas as pd
import scipy.stats as stats
#Chi-square Subject Category Bias in Approvals
# Load the data
train_data = pd.read_csv('train_data.csv')

# Extract relevant columns
data = train_data[['project_subject_categories', 'project_is_approved']]

# Handle missing values if any
data.dropna(subset=['project_subject_categories', 'project_is_approved'], inplace=True)

# Create a contingency table
contingency_table = pd.crosstab(data['project_subject_categories'], data['project_is_approved'])
print("Contingency Table:")
print(contingency_table)

# Perform Chi-square test
chi2, p, dof, ex = stats.chi2_contingency(contingency_table)
print(f"Chi-square statistic: {chi2}")
print(f"p-value: {p}")
print(f"Degrees of freedom: {dof}")

# Interpret the results
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: There is a significant bias in approval rates across different subject categories.")
else:
    print("Fail to reject the null hypothesis: There is no significant bias in approval rates across different subject categories.")

# Identify categories with unexpected approval skews
# Calculate the proportion of approved projects in each category
category_proportions = contingency_table.apply(lambda x: x / x.sum(), axis=1)
print("\nProportion of approved projects in each category:")
print(category_proportions)

#Chi-square statistic: 512.4788727061738
#p-value: 5.942141163769452e-78
#Degrees of freedom: 50
#Chi-square Subject Category Bias in Approvals we can Reject the null hypothesis: There is a significant bias in approval rates across different subject categories.
```

```
import pandas as pd
import scipy.stats as stats
#	Resource Breakdown and Approval
# Load the data
train_data = pd.read_csv('train_data.csv')
resources = pd.read_csv('resources.csv')

# Convert all entries in the 'description' column to strings
resources['description'] = resources['description'].astype(str)

# Aggregate resource descriptions by project ID
resource_categories = resources.groupby('id')['description'].apply(lambda x: ', '.join(set(x))).reset_index()

# Merge the aggregated resource descriptions with train_data
train_data = train_data.merge(resource_categories, on='id', how='left')

# Handle missing values in the new description column
train_data['description'].fillna('Unknown', inplace=True)

# Convert descriptions to categorical data for the Chi-square test
# Create a contingency table
contingency_table = pd.crosstab(train_data['description'], train_data['project_is_approved'])
print("Contingency Table:")
print(contingency_table)

# Perform Chi-square test
chi2, p, dof, ex = stats.chi2_contingency(contingency_table)
print(f"Chi-square statistic: {chi2}")
print(f"p-value: {p}")
print(f"Degrees of freedom: {dof}")

# Interpret the results
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: There is a significant dependence between resource descriptions and project approval.")
else:
    print("Fail to reject the null hypothesis: There is no significant dependence between resource descriptions and project approval.")


#Chi-square statistic: 67544.14492041527
#p-value: 4.8737430048440605e-155
#Degrees of freedom: 58042
#After performing chi square test on Resource Breakdown and Approval we Reject the null hypothesis: There is a significant dependence between resource descriptions and project approval.
```

```
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import xgboost as xgb
import lightgbm as lgb

# Load data
df = pd.read_csv('train_data.csv')

# Combine essay columns and resource summary into one text feature
df['combined_text'] = df[['project_essay_1', 'project_essay_2', 'project_essay_3', 'project_essay_4', 'project_resource_summary']].fillna('').agg(' '.join, axis=1)

# Define feature columns and target
feature_cols = ['project_grade_category', 'project_subject_categories', 'project_subject_subcategories', 'teacher_number_of_previously_posted_projects']
target_col = 'project_is_approved'

# Separate features and target
X = df[feature_cols + ['combined_text']]
y = df[target_col]

# Define preprocessing for numerical and categorical features
numerical_features = ['teacher_number_of_previously_posted_projects']
categorical_features = ['project_grade_category', 'project_subject_categories', 'project_subject_subcategories']

# Numerical pipeline
numerical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# Categorical pipeline
categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Full preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, numerical_features),
        ('cat', categorical_pipeline, categorical_features),
        ('text', TfidfVectorizer(stop_words='english', max_features=5000), 'combined_text')
    ],
    remainder='passthrough'
)

# Define models
models = {
    'Logistic Regression': Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', LogisticRegression(max_iter=1000))
    ]),
    'Naive Bayes': Pipeline([
        ('preprocessor', ColumnTransformer([
            ('text', TfidfVectorizer(stop_words='english', max_features=5000), 'combined_text')
        ], remainder='passthrough')),
        ('classifier', MultinomialNB())
    ]),
    'Decision Tree': Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', DecisionTreeClassifier(random_state=42))
    ]),
    'Random Forest': Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
    ]),
    'SVM': Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', SVC(kernel='linear', probability=True, random_state=42))
    ]),
    'XGBoost': Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42))
    ]),
    'LightGBM': Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', lgb.LGBMClassifier(random_state=42))
    ])
}

# Define stratified k-fold cross-validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Evaluate each model using cross-validation with detailed error messages
results = {}
for model_name, model_pipeline in models.items():
    print(f"Evaluating {model_name}...")
    try:
        cv_results = cross_val_score(model_pipeline, X, y, cv=skf, scoring='accuracy', error_score='raise')
        results[model_name] = cv_results
        print(f"{model_name} Mean CV Accuracy: {cv_results.mean()}  {cv_results.std()}")
    except Exception as e:
        print(f"{model_name} failed: {e}")

# Print summary of results
for model_name, cv_results in results.items():
    if model_name in results:
        print(f"{model_name} Mean CV Accuracy: {cv_results.mean()}  {cv_results.std()}")
```
